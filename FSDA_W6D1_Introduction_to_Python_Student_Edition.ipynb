## **1. Konsep Dasar Bunga Sederhana**

---

Menurut [Pusat Data Kontan](https://pusatdata.kontan.co.id/bungadeposito/), suku bunga deposito saat ini berkisar antara 2% hingga 5% per tahun.

Sebagai contoh, apabila Brian mendepositokan Rp10.000.000,- selama 1 tahun dengan suku bunga tahunan 3,5%, maka Brian akan mendapatkan bunga sebesar Rp350.000, dihitung dengan perhitungan _bunga sederhana_ berikut

$$
\text{bunga} = 0.035 \times 10,000,000 = 350,000
$$

Sehingga setelah 1 tahun, uang Brian bertambah menjadi $\text{saldo_akhir} = 10,000,000+350,000=10,350,000$.


### **Latihan 1.1 | Variabel**

Kemudian gunakan dua variabel itu untuk menghitung dua variabel lainnya:
- `bunga`: Bunga yang didapat Brian setelah 1 tahun
- `saldo_akhir`: Nilai uang Brian setelah 1 tahun

Jangan lupa `print` nilai `bunga` dan `saldo_akhir`.

````python
saldo_awal = 10000000
suku_bunga = 0.035

bunga = suku_bunga * saldo_awal
saldo_akhir = saldo_awal + bunga

print("Bunga yang didapat Brian:", bunga)
print("Saldo akhir setelah 1 tahun:", saldo_akhir)
````

## **2. Konsep Bunga Majemuk / Bunga Berbunga**

---

Setelah 1 tahun, Brian memiliki saldo akhir Rp10.350.000,-. Brian kemudian memutuskan untuk memperpanjang depositonya 1 tahun lagi, dengan suku bunga yang sama 3.5%.

Dengan demikian, setelah 1 tahun lagi, Brian akan mendapatkan bunga dari deposito yang sudah berbunga:

$$
0.035\times 10,350,000=362,250
$$

Sehingga saldo akhir Brian setelah tahun kedua adalah $10,350,000+362,250=10,712,250$.

Konsep bunga berbunga ini juga dikenal sebagai _bunga majemuk (compound interest)_.

Sebagai contoh, jika dihitung dari tahun pertama, saldo akhir Brian di tahun kedua adalah

$$
(1+0.035)^2 \times 10,000,000 = 10,712,250
$$

Di Python, operasi pangkat 2 ditulis dengan dua tanda bintang `**`, contohnya: `(1+0.035)**2`

### **Rumus Saldo Akhir**
Secara umum, rumus dari saldo akhir setelah $n$ periode, dimana setiap periodenya kita mendapat suku bunga $r$ adalah

$$
\text{Saldo akhir} = \text{Saldo awal} \times (1+r)^n
$$

### **Latihan 2.1 | Variabel**

Hitung saldo akhir Brian setelah 2 tahun dengan Rumus Saldo Akhir, lalu masukkan hasilnya ke variabel baru `saldo_akhir_2thn`.

````python
saldo_akhir = saldo_awal * (1+0.035)**2
print(saldo_akhir)
round(saldo_akhir,2)
````

## **3. Pajak Penghasilan Atas Bunga Deposito**

---

Brian adalah WNI dan wajib pajak di Indonesia. Menurut laman [Direktorat Jenderal Pajak](https://www.pajak.go.id/id/pph-pasal-4-ayat-2), bunga deposito yang memiliki saldo awal di atas Rp7.500.000,- akan dikenakan pajak penghasilan (PPh) pasal 4 ayat 2 sebesar 20% setiap tahunnya.

Kembali ke contoh Brian di paling atas, apabila Brian mendepositokan Rp10.000.000,- selama 1 tahun dengan suku bunga tahunan 3,5%, maka Brian terkena PPh dengan tarif 20% (karena saldo awalnya di atas Rp7.500.000,-).

Dengan demikian, sebetulnya Brian akan mendapatkan bunga kotor sebesar Rp350.000, dihitung dengan perhitungan bunga sederhana berikut

$$0.035Ã—10,000,000=350,000$$

Kemudian bunga ini dipotong pajak 20% setiap tahunnya, sehingga bunga bersih yang diterima Brian adalah

$$(1-20\%)\times 350,000= 280,000$$

Dan saldo akhir Brian di akhir tahun akan menjadi Rp10.280.000,- setelah dipotong pajak.

Sebaliknya, apabila Brian hanya mendepositokan uang sebesar Rp5.000.000,- (kurang dari Rp7,5 juta), maka ia tidak akan terkena PPh (atau dengan kata lain, tarif pajaknya 0%).

### **Rumus Saldo Akhir Setelah Pajak (1 periode)**
Secara umum, rumus dari saldo akhir setelah 1 periode, dimana setiap periodenya kita mendapat suku bunga $r$ dan tarif pajak $t$ adalah

$$
\text{Saldo akhir} = \text{Saldo awal} \times (1+(1-t)\times r)
$$

## **4. Saldo Akhir Setelah Pajak Dengan Bunga Majemuk**

---

Setelah adanya pajak, perhitungan saldo akhir pada bunga majemuk menjadi sedikit lebih rumit, dan harus dilakukan setiap tahunnya dengan rumus 1 periode. Hal ini karena tarif pajak yang berlaku mungkin berubah dari tahun ke tahun.

Sebagai contoh di tabel di bawah ini, deposito dengan saldo awal Rp7juta dan suku bunga 10% pada tahun pertama tidak kena pajak, namun di tahun kedua terkena pajak 20%. Hal ini karena saldo awal di tahun kedua bernilai di atas Rp7,5juta.

### **Latihan 4.1 | FOR loop**

Brian membuat kode Python untuk menghitung saldo akhir depositonya setelah 5 tahun. Saldo awal deposito Brian adalah Rp7.000.000 dengan suku bunga 10%/tahun (sama dengan tabel di atas).


```python
# Kode buatan Brian (saldo + pajak)
periode = 5

# Kondisi tahun pertama
saldo_awal_tahun = saldo_awal

# Untuk setiap tahun:
for tahun in range(periode):
  # 1. Tentukan tarif pajak yang berlaku di tahun tersebut
  if saldo_awal_tahun < 7500000:
    tarif_pajak = 0
  elif saldo_awal_tahun >= 7500000:
    tarif_pajak = 0.2

  # 2. Hitung saldo akhir di tahun tersebut dengan Rumus Saldo Akhir Setelah Pajak
  saldo_akhir = saldo_awal_tahun * (1 + (1-tarif_pajak)*suku_bunga)

  # 3. Tentukan saldo awal untuk tahun berikutnya
  saldo_awal_tahun = saldo_akhir

print(f'Saldo akhir sesudah pajak : {saldo_akhir:,.0f}')
   
```

Apakah kode buatan Brian benar? Jika tidak, perbaiki kode buatan Brian (hint: Saldo akhir Brian, `saldo_akhir`, seharusnya sekitar Rp10.475.765)

````python
# Kode buatan Brian (saldo + pajak)
periode = 5
saldo_awal = 7000000
suku_bunga = 0.1

# Kondisi tahun pertama
saldo_awal_tahun = saldo_awal
saldo_akhir = 0
tarif_pajak = 0

# Untuk setiap tahun:
for tahun in range(periode):
  # 1. Tentukan tarif pajak yang berlaku di tahun tersebut
  if saldo_awal_tahun < 7500000:
    tarif_pajak = 0
  elif saldo_awal_tahun >= 7500000:
    tarif_pajak = 0.2

  # 2. Hitung saldo akhir di tahun tersebut dengan Rumus Saldo Akhir Setelah Pajak
  saldo_akhir = saldo_awal_tahun * (1 + (1-tarif_pajak)*suku_bunga)

  # 3. Tentukan saldo awal untuk tahun berikutnya
  saldo_awal_tahun = saldo_akhir

print(f'Saldo akhir sesudah pajak : {saldo_akhir:,.0f}')
````
